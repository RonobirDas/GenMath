```python
s = fourier_series(Heaviside(t) - Heaviside(t-1), (t, 0, 2))
s.truncate(4)
```

[Source](https://stackoverflow.com/questions/36177131/sympy-extract-fourier-series-coefficients)

```python
iter = s.truncate(None)
cosine_coeffs = []
sine_coeffs = [0]       # there is no sine term for k = 0
amplitude = []
phase = []
for k in range(0, 4):
    term = next(iter)
    #print(term)
    cosine_coeffs.append(term.subs(t, 0))
    if k > 0:
        sine_coeffs.append(term.subs(t, k/2)) #correction, we only care about 1/2 * pi terms since sin(n*pi/2) = 1, which is what we care about
for l in range(len(sine_coeffs)):
    amplitude.append(sqrt((sine_coeffs[l])**2 + (cosine_coeffs[l])**2))
    phase.append(-atan2(sine_coeffs[l],cosine_coeffs[l]))
    print(phase[l])
```
